```python
import pandas as pd 
import matplotlib as mpl 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
data = pd.read_csv('mphdropspin.csv')
```

```python
print(data.columns)
```

```python
ffmph = np.array(data.ff_avg_speed)
ffspin = np.array(data.ff_avg_spin)
ffbreak = np.array(data.ff_avg_break_z)
```

```python
x = ffmph
y = ffspin
```

```python
print(x)
print(y)
```

```python
plt.scatter(x, y)
plt.xlabel('Velocity(MPH)')
plt.ylabel('Spin Rate(RPM)')
plt.title('Comparing Fastball Velocity and Spin Rate in RPMs.')
plt.show
```
